```python
import math


def refactor_col(annotations_df, col_name, convert = 1):
    '''
    Function used for changing cloumn types of annotaions df

    :param annotations_df (DataFrame): global annptaions dataframe 
    :param col_name (String): name of column to refactor

    '''
    new_col = []
    for row in annotations_df[col_name]:

        if np.all(pd.isnull(row)):
            new_col.append(np.nan)
            continue
        elif row == "Not active":
            print("not active")
            new_col.append(row)
            continue
        else:
            if convert == 1:
                temp_list = list(row.split(","))
                for num in range(len(temp_list)):
                    temp_list[num] = float(temp_list[num])
                new_col.append(temp_list)
            else:
                temp_list = list(row.split(","))
                for num in range(len(temp_list)):
                    temp_list[num] = int(temp_list[num])
                new_col.append(temp_list)
    new_df = pd.Series(new_col)
    return new_df
#     return new_col
```

```python
data_temp = bp_data
activity_cols = bp_data.columns.values[3:19]
for o in activity_cols:
    if (bp_data.dtypes[o] == np.float):
        print("col in float")
        continue
    else:
        bp_data[o] = refactor_col(data_temp, o)
bp_data
```
