```python
def refactor_col_delim(annotations_df, col_name):
    '''
    Function used for changing cloumn types of annotaions df

    :param annotations_df (DataFrame): global annptaions dataframe 
    :param col_name (String): name of column to refactor

    '''
    new_col = []
    for row in annotations_df[col_name]:

        if np.all(pd.isnull(row)):
            new_col.append(np.nan)
            continue
        elif row == "Not active":
            print("not active")
            new_col.append(row)
            continue
        else:
            temp_list = row.split(",")
            for num in range(len(temp_list)):
                temp_list[num] = str(float(temp_list[num]))
            new_col.append("|".join(temp_list))
    
    new_df = pd.Series(new_col)
    return new_df


def refactor_col_delim_heat(annotations_df, col_name):
    '''
    Function used for changing cloumn types of annotaions df

    :param annotations_df (DataFrame): global annptaions dataframe 
    :param col_name (String): name of column to refactor

    '''
    new_col = []
    for row in annotations_df[col_name]:

        if np.all(pd.isnull(row)):
            new_col.append(np.nan)
            continue
        else:
            temp_list = row.split(",")
            for num in range(len(temp_list)):
                temp_list[num] = str(int(temp_list[num]))
            new_col.append("|".join(temp_list))
    
    new_df = pd.Series(new_col)
    return new_df
```

```python
temp_d = df
cols = df.columns.values[3:19]
cols
for o in cols:
    if (df.dtypes[o] == np.float):
        print("col in float")
        continue
    else:
        df[o] = refactor_col_delim(temp_d, o)
df
```

```python
temp_d = df
cols = df.columns.values[27:]
cols
for o in cols:
    print(o)
    df[o] = refactor_col_delim_heat(temp_d, o)
df
```

```python
new_cvs = "orig_annotations_060420.csv"
new_data = 'orig_refactored_annotations.txt'
df.to_csv(new_cvs, mode='w', index=False)

csv_file2 = new_cvs
txt_file2 = new_data

with open(txt_file2, "a") as my_output_file:
    with open(csv_file2, "r") as my_input_file:
        [ my_output_file.write("\t".join(row)+'\n') for row in csv.reader(my_input_file)]
    my_output_file.close()


```
